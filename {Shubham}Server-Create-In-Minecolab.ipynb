```python
# Creating Backup for Not To loss data.


# Navigate to the Minecraft server directory
%cd /content/drive/MyDrive/MinecraftServer

import os
import shutil
from datetime import datetime
import pytz  # Import pytz for timezone handling

# Define the path to your Minecraft server and backup location
server_directory = '/content/drive/MyDrive/MinecraftServer'  # Change this to your server's directory
backup_directory = '/content/drive/MyDrive/minecraft_backups'

# Specify your local timezone (e.g., 'America/New_York', 'Europe/London', 'Asia/Kolkata')
local_timezone = 'Asia/Kolkata'  # Replace with your local timezone

# Function to create a zip backup of the server directory
def create_backup_zip(server_directory, backup_directory):
    if not os.path.exists(backup_directory):
        os.makedirs(backup_directory)

    # Get the current local time
    local_time = datetime.now(pytz.timezone(local_timezone))

    # Create a timestamp with parentheses for better readability
    timestamp = local_time.strftime("(%Y,%m,%d,%H,%M,%S)")
    backup_zip_path = os.path.join(backup_directory, f'minecraft_backup_{timestamp}.zip')

    # Create the zip file
    shutil.make_archive(backup_zip_path.replace('.zip', ''), 'zip', server_directory)
    return backup_zip_path

# Create a backup zip file
print("Creating a backup of the server files...")
backup_zip = create_backup_zip(server_directory, backup_directory)

print(f"Backup completed! File saved to: {backup_zip}")
```
